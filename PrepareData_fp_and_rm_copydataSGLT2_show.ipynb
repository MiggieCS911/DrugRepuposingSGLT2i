```python
df =df.drop_duplicates(subset=['molecule_chembl_id'], keep = 'first')
```

```python
df = df.reset_index()
```

```python
#check activity class
df.bioactivity.value_counts()
```

```python
# create df from canonical_smiles to pubmedfingerprint
from padelpy import from_smiles
```

# Creat dataframe with fp from canonical smiles

```python
#create dataframe that contain pubchem fp

c_smi = df.canonical_smiles.to_list() #turn to list to be input

i = 0
while i < len(c_smi):
  length = 5
  if i == 0:
    temp = c_smi[i: i+length]
    pubchem = from_smiles(temp, fingerprints=True, descriptors=False )
    newdf = pd.DataFrame(pubchem)
  elif i +10 > len(c_smi):
    length = len(c_smi) - i
    temp = c_smi[i: i+length]
    pubchem = from_smiles(temp, fingerprints=True, descriptors=False )
    tempdf = pd.DataFrame(pubchem)
    newdf = pd.concat([newdf,tempdf], axis=0, ignore_index=True)
  else:
    temp = c_smi[i:i+length]
    pubchem = from_smiles(temp, fingerprints=True, descriptors=False )
    tempdf = pd.DataFrame(pubchem)
    newdf = pd.concat([newdf,tempdf], axis=0, ignore_index=True)
  i+=length 
  print(i)

# pubchem = from_smiles(c_smi, fingerprints=True, descriptors=False )

newdf.shape
```
